```python
#fa4prog2

password = "@sixseven"

max_tries = 3
attempts_made = 0

while True:
  entered_password = input("Please enter your password: ")

  if entered_password == password:
    print("Access Granted!")
    break

  else:
    attempts_made += 1
    print("Incorrect Password.")

    if attempts_made == max_tries:
      print("Maximum Attempts Reached. Access Denied.")
      break

    else:
      print(f"Attempts remaining: {max_tries - attempts_made}")
```
